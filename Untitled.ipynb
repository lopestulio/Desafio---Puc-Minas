```python
import pandas as pd
import geopandas as gpd
import matplotlib as plt
import matplotlib.pyplot as plt
import scipy.stats as stats
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python

df = pd.read_excel("C:\\Users\\lopes\\OneDrive\\Área de Trabalho\\Desafio\\Nova pasta\\analise_por_raca.xlsx")


df.head()
```

```python

vars_1 = ['preta_parda', 'educacao', 'IPM']
vars_2 = ['preta_parda', 'saude', 'IPM']

# Normalização dos conjuntos de dados
scaler = StandardScaler()
df_normalized_1 = scaler.fit_transform(df[vars_1])
df_normalized_2 = scaler.fit_transform(df[vars_2])


df_normalized_1[:5], df_normalized_2[:5]
```

```python
import scipy.cluster.hierarchy as sch
import numpy as np

# Função para plotar dendrograma
def plot_dendrogram(model, **kwargs):
    # Cria os counts das amostras sob o nó
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # folha
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)
    # Plota o dendrograma correspondente
    sch.dendrogram(linkage_matrix, **kwargs)

# Aplicando agrupamento hierárquico
from sklearn.cluster import AgglomerativeClustering
model_1 = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model_1 = model_1.fit(df_normalized_1)

# Plotando o dendrograma para o Conjunto 1
plt.figure(figsize=(10, 7))
plt.title('Dendrograma para o Conjunto 1')
plot_dendrogram(model_1, truncate_mode='level', p=3)
plt.xlabel("Número de pontos no nó (ou índice do ponto se não houver parênteses).")
plt.show()
```

```python

model_2 = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model_2 = model_2.fit(df_normalized_2)


plt.figure(figsize=(10, 7))
plt.title('Dendrograma para o Conjunto 2')
plot_dendrogram(model_2, truncate_mode='level', p=3)
plt.xlabel("Número de pontos no nó (ou índice do ponto se não houver parênteses).")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Aplicando DBSCAN aos conjuntos de dados normalizados
dbscan_model_1 = DBSCAN(eps=0.5, min_samples=5).fit(df_normalized_1)
dbscan_model_2 = DBSCAN(eps=0.5, min_samples=5).fit(df_normalized_2)

# Adicionando os rótulos dos clusters aos dados originais
df['Cluster_1_DBS'] = dbscan_model_1.labels_
df['Cluster_2_DBS'] = dbscan_model_2.labels_

# Mostrando os primeiros registros com os rótulos dos clusters
df[['preta_parda', 'educacao', 'IPM', 'Cluster_1_DBS', 'saude', 'Cluster_2_DBS']].head()
```

```python
# Contagem da distribuição dos rótulos dos clusters
cluster_distribution_1 = df['Cluster_1_DBS'].value_counts()
cluster_distribution_2 = df['Cluster_2_DBS'].value_counts()

cluster_distribution_1, cluster_distribution_2
```

```python
from sklearn.decomposition import PCA

# Aplicando PCA para redução de dimensionalidade
pca = PCA(n_components=2)
principal_components_1 = pca.fit_transform(df_normalized_1)
principal_components_2 = pca.fit_transform(df_normalized_2)


df['PC1_1'] = principal_components_1[:, 0]
df['PC2_1'] = principal_components_1[:, 1]
df['PC1_2'] = principal_components_2[:, 0]
df['PC2_2'] = principal_components_2[:, 1]

# Plotando os clusters 
plt.figure(figsize=(10, 7))
plt.scatter(df['PC1_1'], df['PC2_1'], c=df['Cluster_1_DBS'], cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('Visualização dos Clusters para o Conjunto 1')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar()
plt.show()
```

```python
# Plotando os clusters 
plt.figure(figsize=(10, 7))
plt.scatter(df['PC1_2'], df['PC2_2'], c=df['Cluster_2_DBS'], cmap='rainbow', alpha=0.7, edgecolors='b')
plt.title('Visualização dos Clusters para o Conjunto 2')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar()
plt.show()
```

```python
# Definindo os limites para a variável categórica baseada no IPM
ipm_min, ipm_max = df['IPM'].min(), df['IPM'].max()
ipm_range = ipm_max - ipm_min

# Criando a variável categórica com três níveis
df['IPM_Categoria'] = pd.cut(df['IPM'], bins=[ipm_min, ipm_min + ipm_range/3, ipm_min + 2*ipm_range/3, ipm_max], labels=['Baixo', 'Médio', 'Alto'], include_lowest=True)

# Mostrando a distribuição dos níveis da variável categórica após a alteração
ipm_category_distribution_updated = df['IPM_Categoria'].value_counts()

ipm_category_distribution_updated
```

```python
# Criando variáveis dummy para a variável "IPM_Categoria"
ipm_dummy = pd.get_dummies(df['IPM_Categoria'], prefix='IPM')


df_with_dummies = pd.concat([df, ipm_dummy], axis=1)


variables_for_clustering = ['preta_parda', 'educacao', 'IPM_Baixo', 'IPM_Médio', 'IPM_Alto']
df_for_clustering = df_with_dummies[variables_for_clustering]


scaler = StandardScaler()
df_for_clustering_scaled = scaler.fit_transform(df_for_clustering)


df_for_clustering_scaled[:5,:]
```

```python


k_values = range(1, 11)


sse = [KMeans(n_clusters=k, random_state=42).fit(df_for_clustering_scaled).inertia_ for k in k_values]

# Criando o gráfico do método do cotovelo
plt.figure(figsize=(10, 5))
plt.plot(k_values, sse, marker='o')
plt.title('Método do Cotovelo para Escolher o Número de Clusters')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Soma dos Quadrados das Distâncias')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
kmeans_model = KMeans(n_clusters=3, random_state=42)
df['Cluster_KMeans'] = kmeans_model.fit_predict(df_for_clustering_scaled)


cluster_kmeans_distribution = df['Cluster_KMeans'].value_counts()

cluster_kmeans_distribution
```

```python
# Criando um DataFrame com os resultados do K-means e os componentes principais
df_plot = df.copy()
df_plot['PC1'], df_plot['PC2'], df_plot['PC3'] = df_for_clustering_scaled[:,0], df_for_clustering_scaled[:,1], df_for_clustering_scaled[:,2]

# Criando o gráfico 3D
fig = px.scatter_3d(df_plot, x='PC1', y='PC2', z='PC3',
                    color='Cluster_KMeans', text='nom_mun',
                    labels={'Cluster_KMeans': 'Cluster'},
                    title='Visualização 3D dos Clusters K-means')
fig.update_layout(width=800, height=600)
# Mostrando o gráfico
fig.show()
```
